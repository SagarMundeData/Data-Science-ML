```python
#Allocating data to x and y for train test split
X = df['Sqft'].values.reshape(-1,1)
y = df['Price'].values.reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=0)
```
